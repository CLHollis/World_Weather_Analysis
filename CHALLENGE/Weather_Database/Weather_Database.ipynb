#### 1.3 Create a new set of 2,000 random latitudes and longitudes

```python
# generates latitudes and longitudes & declare each array as a variable
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)

# pack the latitudes (lats) and longitudes (lngs) as pairs by 
#  zipping them (lat_lngs) with the zip() function.
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
# Add the latitudes and longitudes to a list.
coordinates = list(lat_lngs)
```

#### 1.4 Get the nearest city using the citipy module.

```python
# import the citipy script from the citipy module
from citipy import citipy
```

```python
# Create a cities list to store city names.
cities = []

# iterate through the coordinates and retrieve the nearest city using the latitude and longitude pair
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    # add a decision statement with the logical operator "not" in 
    # to determine whether the found city is already in the cities list
    if city not in cities:
        # If not, then we'll use the append() function to add it
        cities.append(city)
        
        # We are doing this because among the 1,500 latitudes and longitudes, there might be duplicates, 
        # which will retrieve duplicate cities, and we want to be sure we capture only the unique cities.
        
        # Print the city count to confirm sufficient count.
len(cities)
```

#### 1.6 Retrieve the following information from the API call:
- Latitude and longitude
- Maximum temperature
- Percent humidity
- Percent cloudiness
- Wind speed
- Weather description (for example, clouds, fog, light rain, clear sky)

```python
# Create an empty list to hold the weather data.
city_data = []

# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
# initialized the counters at 1 because we want the first iteration of the logging 
# ...for each recorded response and the set to start at 1
record_count = 1
set_count = 1

# Loop through all the cities in our list.
# Refrence index (i) & city in list
for i, city in enumerate(cities):

    # Group cities in sets of 50 for logging purposes.
    # (if the remainder of i/50=0 AND i>=50)
    if (i % 50 == 0 and i >= 50):
        # If true, then...
        set_count += 1
        record_count = 1
        time.sleep(60)

    # Create endpoint URL with each city 
    # ...& removing the blank spaces and concatenating the city name
    city_url = url + "&q=" + city.replace(" ","+")

    # print statement that tells us the record count & set count, & the city being processed
    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
   
    # Add 1 to the record count.
    record_count += 1
    
    # Run an API request for each of the cities.
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
        
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_description = city_weather["weather"][0]["description"]
        
        # Create Column Headers and set value to their coresponding variable from above
        # Append the city information into city_data list.
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Current Description": city_description})

    # If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

#### 1.7 Add the data to a new DataFrame

```python
# Convert the array of dictionaries to a Pandas DataFrame.
city_data_df = pd.DataFrame(city_data)

# print top 10 rows only
city_data_df.head(10)
```

```python
# Assign a variable to an array of the columns in the order we want them to appear:
new_column_order = ["City", "Country", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Current Description"]

# Assign a new or the same DataFrame with new column order:
city_data_df = city_data_df[new_column_order]

# print
city_data_df.head(10)
```

#### 1.8 Export the DataFrame as a CSV file, and save it as WeatherPy_Database.csv in the Weather_Database folder.

```python
# Create the output file (CSV) called WeatherPy_Database.
output_data_file = "WeatherPy_Database.csv"

# Export the City_Data into a CSV.
city_data_df.to_csv(output_data_file, index_label="City_ID")
```
